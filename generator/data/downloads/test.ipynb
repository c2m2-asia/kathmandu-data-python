```python
def update_null_how_long(target_col, value_col):
    target_col = target_col.values.tolist()
    value_col = value_col.values.tolist()
    output = []
    for idx, i in enumerate(target_col):
        if i == -1:
            value = value_col[idx]
        else:
            value = i
        output.append(value)
    return output
```

```python
business_impact['how_lng_stop_business'] = update_null_how_long(business_impact['how_lng_stop_business'], business_impact['stop_business'].apply(lambda x: x.split(', ')[1].capitalize()))
```
